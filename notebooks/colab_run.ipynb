# cad-cms — Colab runner

Thin wrapper: clone the repo, install dependencies, mount Drive for the
MVTec AD data, point the config at that data, and run the benchmark.

```python
REPO_URL = "https://github.com/Navs17/cad-cms.git"
```

```python
!git clone $REPO_URL cad-cms
%cd cad-cms
!pip install -q -r requirements.txt
```

```python
from google.colab import drive
drive.mount('/content/drive')

# Point this at wherever MVTec AD lives in your Drive, laid out as described
# in data/README.md (data/MVTec-AD/<category>/{train,test,ground_truth}).
DRIVE_DATA_ROOT = "/content/drive/MyDrive/MVTec-AD"
```

```python
import yaml

with open("configs/default.yaml") as f:
    config = yaml.safe_load(f)

config["paths"]["data_root"] = DRIVE_DATA_ROOT
config["data"]["num_workers"] = 2  # safe to raise on Colab, unlike Windows default of 0

with open("configs/colab.yaml", "w") as f:
    yaml.safe_dump(config, f)
```
